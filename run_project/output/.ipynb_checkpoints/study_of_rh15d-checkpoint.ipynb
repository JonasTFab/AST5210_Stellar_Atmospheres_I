# Analysis of RH 1.5D code

----------

First we want to look at the atmospheric data to locate various features. Note that we are looking at a coronal hole generated from the Bifrost code (currently snapshot 800, image 42(?)). 

```python
#%matplotlib ipympl
import h5py, numpy as np, matplotlib.pyplot as plt

file_path = "../../../Atmos/bifrost_ch012012_hion_snap800.hdf5"
file = h5py.File(file_path, "r")
```

```python
steps = np.linspace(0, len(file["z"][0])-1, len(file["z"][0]))

plt.figure(figsize=(16,5))

plt.subplot(121)
plt.title("z")
plt.plot(steps, file["z"][0])


plt.subplot(122)
plt.title("z")
plt.plot(steps, file["z"][0])
plt.yscale("log")

plt.tight_layout()
```

```python
circ = (100,100)
circ_col = "red"
z_num = 500

plt.figure(figsize=(16,5))

plt.subplot(131)
plt.title("electron_density (z=%i)" % z_num)
plt.contourf(file["electron_density"][0,:,:,z_num], cmap="gray")
circle1 = plt.Circle(circ, 50, color=circ_col, fill=False, clip_on=False, label="Pixel=(%i,%i)" % (circ[0], circ[1]))
plt.gca().add_patch(circle1)
plt.xlabel("Pixel"); plt.ylabel("Pixel")
plt.colorbar()
plt.legend()


plt.subplot(132)
plt.title("temperature (z=%i)" % z_num)
plt.contourf(file["temperature"][0,:,:,z_num], cmap="gray")
circle1 = plt.Circle(circ, 50, color=circ_col, fill=False, clip_on=False, label="Pixel=(%i,%i)" % (circ[0], circ[1]))
plt.gca().add_patch(circle1)
plt.xlabel("Pixel"); plt.ylabel("Pixel")
plt.colorbar()
plt.legend()


plt.subplot(133)
plt.title("velocity_z (z=%i)" % z_num)
plt.contourf(file["velocity_z"][0,:,:,z_num], cmap="gray")
circle1 = plt.Circle(circ, 50, color=circ_col, fill=False, clip_on=False, label="Pixel=(%i,%i)" % (circ[0], circ[1]))
plt.gca().add_patch(circle1)
plt.xlabel("Pixel"); plt.ylabel("Pixel")
plt.colorbar()
plt.legend()

plt.tight_layout()
```
