```python
f = open("../ray.input", "w")  # this will overwrite any existing file!
f.write("1.00\n")
#output = str(len(indices))                    # includes all of the Mg II spectrum
#selected_indices = indices                    # includes all of the Mg II spectrum
output = str(len(idx_k))                       # only Mg II k profile
selected_indices = int(min(indices)) + idx_k   # only Mg II k profile
for ind in selected_indices:
    output += " %i" % ind
f.write(output)
f.close()
```

```python
mg_k = [279.45+vac, 279.66+vac]
mg_h = [280.17+vac, 280.38+vac]
idx_k = np.arange(len(wave))[(wave > mg_k[0]) & (wave < mg_k[1])]
idx_h = np.arange(len(wave))[(wave > mg_h[0]) & (wave < mg_h[1])]

I = data.ray.intensity[col_idx[0], col_idx[1]][indices]
I_k = I[idx_k]       # intensity around k line
I_h = I[idx_h]       # intensity around h line



wave_k = wave[idx_k]
idx_k_centre = np.where(np.min(abs(wave_k-wl_k_rest)) == abs(wave_k-wl_k_rest))[0]

while True:
    I_r = I_k[idx_k_centre+1][0]     # right of temporary centre
    I_l = I_k[idx_k_centre-1][0]     # left of temporary centre
    
    if I_r < I_k[idx_k_centre]:
        idx_k_centre += 1
    elif I_l < I_k[idx_k_centre]:
        idx_k_centre -= 1
    else:
        break
wl_k_centre = wave_k[idx_k_centre][0]



wave_h = wave[idx_h]
idx_h_centre = np.where(np.min(abs(wave_h-wl_h_rest)) == abs(wave_h-wl_h_rest))[0]

while True:
    I_r = I_h[idx_h_centre+1][0]     # right of temporary centre
    I_l = I_h[idx_h_centre-1][0]     # left of temporary centre
    
    if I_r < I_h[idx_h_centre]:
        idx_h_centre += 1
    elif I_l < I_h[idx_h_centre]:
        idx_h_centre -= 1
    else:
        break
wl_h_centre = wave_h[idx_h_centre][0]


from astropy import constants, units

c = constants.c.to("nm/s")
dnu_k = (wl_k_rest - wave_k) / wl_k_rest * c * 1e-12                # converting to correct units (km/s)
dnu_h = (wl_h_rest - wave_h) / wl_h_rest * c * 1e-12                # converting to correct units (km/s)
dnu_k_centre = (wl_k_rest - wl_k_centre) / wl_k_rest * c * 1e-12    # converting to correct units (km/s)
dnu_h_centre = (wl_h_rest - wl_h_centre) / wl_h_rest * c * 1e-12    # converting to correct units (km/s)


dnu_lim = 75
min_idx_k = min(idx_k)
min_idx_h = min(idx_h)
idx_k = np.where(abs(dnu_k)<dnu_lim)[0] + min_idx_k     # redefine idx_k
idx_h = np.where(abs(dnu_h)<dnu_lim)[0] + min_idx_h     # redefine idx_h

# redefine parameters originally defined with idx_k and idx_h
dnu_k = dnu_k[idx_k-min_idx_k]
dnu_h = dnu_h[idx_h-min_idx_h]
wave_k = wave[idx_k]
wave_h = wave[idx_h]
I_k = I[idx_k]       # intensity around k line
I_h = I[idx_h]       # intensity around h line
```

```python
plt.figure(figsize=(20,7))
plt.subplot(121)
#plt.plot(wave[idx_k], I_k)
plt.plot(wave_k, I_k)
plt.plot([wl_k_rest, wl_k_rest], [0, 4.5e-9], label="Rest", color="Red")
plt.plot([wl_k_centre, wl_k_centre], [0, 4.5e-9], "--", label="Centre", color="Red")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [W Hz⁻¹ m⁻² sr⁻¹]")
plt.grid()
plt.legend()
plt.title("Mg II k")

plt.subplot(122)
#plt.plot(wave[idx_h], I_h)
plt.plot(wave_h, I_h)
plt.plot([wl_h_rest, wl_h_rest], [0, 4.5e-9], label="Rest", color="Green")
plt.plot([wl_h_centre, wl_h_centre], [0, 4.5e-9], "--", label="Centre", color="Green")
plt.xlabel("Wavelength [nm]")
plt.ylabel("Intensity [W Hz⁻¹ m⁻² sr⁻¹]")
plt.grid()
plt.legend()
plt.title("Mg II h");

dwl_k = wl_k_centre-wl_k_rest
dwl_h = wl_h_centre-wl_h_rest
print("Doppler shift in Mg II k: %.5f nm" % dwl_k)
print("Doppler shift in Mg II h: %.5f nm" % dwl_h)

```

We note the centre is slighlty red-shifted. This means that the atmosphere in this column is moving towards what is defined as the negative direction. The atmosphere from Bifrost simulation has downflow as positive direction which means that the velocity of atmosphere is outgoing.

Converting wavelengths to $\Delta \nu$ [km/s] and replot.
